```python
browser.get(lkf_in)
links = []
#40 sek
```

```python
#loop starts here
content = browser.page_source
#2,5 sek
#1,5 sek
```

```python
soup = bs4.BeautifulSoup(content)
#1 sek
#2 sek
```

```python
grids = soup.select(".gridcell")
#links = []
for item in grids:
    #print(item.find('a'))
    for a in item.find_all('a'):
        if a['href'][:13] == 'ObjectDetails':
            links.append(a['href'])

#0,7 sek
#0,6 sek
#8,6 sek
```

```python
# finding the button using ID
button = browser.find_element_by_id("ctl00_ctl01_DefaultSiteContentPlaceHolder1_Col1_ucNavBar_btnNavNext")

# clicking on the button
button.click()

#1,5 sek
#2,8 sek
#"loop ends here" go through all pages
```

```python
#create list with usable links to each object
objekts = []
for l in links:
    objekts.append('https://minasidor.lkf.se/HSS/Object/'+l)
```

```python
#navigate to object
#loop starts here
browser.get(objekts[50])
#10 sek
#36 o 37 38 39 korttidskontrakt krashar
```

```python
content = browser.page_source

#5 sek
```

```python
soup = bs4.BeautifulSoup(content)
#0,5 sek
```

```python
ruta2 = soup.select("#ctl00_ctl01_DefaultSiteContentPlaceHolder1_Col1_divAriaFacts")
ruta1 = soup.select("#ctl00_ctl01_DefaultSiteContentPlaceHolder1_Col1_divAvailabilityBoxTop")

#0,3 sek
```

```python
data1 = ruta1[0].text
#6 sek
```

```python
datalista1 = data1.split("\n")[3:-2]
```

```python
data2 = ruta2[0].text
```

```python
datalista2 = data2.split("\n")[4:-2]
```

```python
datalista1.extend(datalista2)
```

```python
res = [ele for ele in datalista1 if ele != ""]
```

```python
object_data = {}
for i in range(0,len(res),2):
    object_data[res[i]]=res[i+1]
    
```

```python
#object_data
```

```python
object_list.append(object_data)
```
